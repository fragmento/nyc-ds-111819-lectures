### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.columns
```

```python
# create pivot
gender_user = users.pivot_table(index='occupation', columns=['gender'], fill_value=0, aggfunc='size')
gender_user
```

```python
users['female'] = np.where(users['gender'] == 'F',1,0)
```

```python
users[['occupation','female']]
```

```python
gender_user['sums'] = gender_user[['F', 'M']].sum(axis=1)
gender_user.head()
```

```python


# calculate male ratio
gender_user['male_ratio'] = gender_user['M']/gender_user['sums']
gender_user
sorted_by_gender_ratio = gender_user.sort_values(by = ['male_ratio'])

sorted_by_gender_ratio
```
