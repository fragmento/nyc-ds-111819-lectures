```python
import csv

with open('data.csv') as f:
        # we are using DictReader because we want our information to be in dictionary format.
        reader = csv.DictReader(f)
        my_music = list(reader)
        
```

```python
def by_name(album):
    for v in my_music:
        if album == v['album']:
            return v
        
        
```

```python
def by_artist(artist,music_list): #returns all albums by the same artist
    hits = []
    for v in music_list:
        if artist == v['artist']:
            hits.append(v['album'])
    return hits
    
```

```python
by_artist('The Beatles',my_music)
```

```python
def by_rank(number,music_list):
    for v in music_list:
        if int(number) == int(v['number']):
            return v
        
```

```python
by_rank(17,my_music)
```

```python
def by_year(year,music_list):
    same_year = []
    for v in music_list:
        if int(year) == int(v['year']):
            same_year.append(v)
    return same_year
```

```python
by_year(1982, my_music)
```

```python
def by_years(year1,year2,music_list): #first year has to be smaller than second year
    same_year = []

    for v in music_list:
        if int(year1) <= int(v['year']) and int(year2) >= int(v['year']):
            same_year.append(v)
    return same_year
```

```python
by_years(1966,1967,my_music)
```

```python
def by_ranks(rank1,rank2,music_list): #first rank has to be smaller than second rank
    ranks = []

    for v in music_list:
        if int(rank1) <= int(v['number']) and int(rank2) >= int(v['number']):
            ranks.append(v)
    return ranks
```

```python
by_ranks(1,3,my_music)
```
